```python
df_fin_ratio = pd.read_excel('Project_Work/823_analytics_for_financial_markets/Financial Ratios Data 2003_2023x.xlsm',header=1)
df_stock_return  = pd.read_excel('Project_Work\823_analytics_for_financial_markets\Stock_Return_Data2023Fwd.xlsx')
df_sector_info = pd.read_excel('Project_Work\823_analytics_for_financial_markets\Sector Information.xlsm')
```

```python
ratio_stock_return_merge = df_fin_ratio.merge(df_stock_return, left_on='PERMNO-YearMonth', right_on='PERMNO_DATE_LABEL', how='inner')
```
