```python
#changing the names the monthly payment of the features
df.rename(columns={
'pmtstatus01':'pmtstatus1',
'pmtstatus02':'pmtstatus2',
'pmtstatus03':'pmtstatus3',
'pmtstatus04':'pmtstatus4',
'pmtstatus05':'pmtstatus5',
'pmtstatus06':'pmtstatus6',
'pmtstatus07':'pmtstatus7',
'pmtstatus08':'pmtstatus8',
'pmtstatus09':'pmtstatus9'},
inplace = True
)
```

```python
#Adding dummy variables for gender
df = pd.get_dummies(df, columns=['gender_at_activation'], prefix='gender')

# View the updated DataFrame
print(df)
```

```python
#Adding dummy variables for marital status
df = pd.get_dummies(df, columns=['marital_status_at_activation'], prefix='marital status')
df.columns
```

```python
#Making housing status binary
ownership_mapping = {
    'Own': 1,
    'Rent': 0
}

# Apply the mapping to the 'ownership' column
df['housing_status_at_activation_binary'] = df['housing_status_at_activation'].map(ownership_mapping)

#dropping old column
df.drop('housing_status_at_activation', axis=1, inplace= True)
df.head(10)
```

```python
#dummy variables for employment status
df = pd.get_dummies(df, columns=['employment_status_at_activation'], prefix='emp_type')
df.columns
```

```python
#there is a order to this... i think zero<partial<full... so I will use ordinal encoding...
#I think I could use hot encoding as well but we'll try ordinal first

pmt_status_mapping = {
    'NaN': -1,
    'zero': 0,
    'partial': 1,
    'full': 2
}
# Apply mapping to each payment status column
for i in range(1, 13):
    column_name = f'pmtstatus{i}'
    if column_name in df.columns:
        df[column_name] = df[column_name].map(pmt_status_mapping)
# View the converted DataFrame
df.head(10)
```

```python
#changing all my booleans to integers
df[df.select_dtypes(include='bool').columns] = df.select_dtypes(include='bool').astype(int)
```

```python
#more cleaning, we have to get rid of nans and stuff
nan_counts = df.isna().sum()

# Display columns with missing values
print(nan_counts[nan_counts > 0])
```

```python
df1 = df.drop(['occupation_at_activation', 'reason_for_DCP_at_activation'], axis=1).dropna(subset=['age_at_activation','postal_prefix_at_activation','city_at_activation','province_at_activation','monthly_expenses_at_activation'])
```

```python
#more cleaning, we have to get rid of nans and stuff
nan_counts1 = df1.isna().sum()

# Display columns with missing values
print(nan_counts1[nan_counts1 > 0])
```

```python
df1['pmtstatus12'].head(10)
```

```python
# Define the payment status columns
payment_status_columns = [f'pmtstatus{i}' for i in range(1, 13)]

# Replace only NaN values in the specified columns with -1
df1[payment_status_columns] = df[payment_status_columns].fillna(-1)

# Verify the changes
print(df1[payment_status_columns].head())
```

```python
df1['pmtstatus12'].head(10)
```

```python
#Feature Engineering
#Asset networth
df1['asset_net_woth'] = df1['asset_value_at_activation']-df1['asset_amount_owed_at_activation']

#Expense Income Ratio
df1['expense_income_ratio']=df1['monthly_expenses_at_activation']/df1['monthly_income_at_activation']

df1['debt_income_ratio']=(df1['DCP_debt_balance_at_activation']+df1['asset_amount_owed_at_activation'])/df1['monthly_income_at_activation']
```

```python
df1.info()
```

```python
#lets change all the date time features

#activation
df1['DCP_activation_date_year']=df1['DCP_activation_date'].dt.year
df1['DCP_activation_date_month']=df1['DCP_activation_date'].dt.month
df1['DCP_activation_date_day']=df1['DCP_activation_date'].dt.dayofweek

#start
df1['DCP_start_date_year']=df1['DCP_start_date'].dt.year
df1['DCP_start_date_month']=df1['DCP_start_date'].dt.month
df1['DCP_start_date_day']=df1['DCP_start_date'].dt.dayofweek

#end
df1['DCP_drop_date_year']=df1['DCP_drop_date'].dt.year
df1['DCP_drop_date_month']=df1['DCP_drop_date'].dt.month
df1['DCP_drop_date_day']=df1['DCP_drop_date'].dt.dayofweek


df1.info()
```

```python
df1['activation_month_sin'] = np.sin(2 * np.pi * df1['DCP_activation_date_month'] / 12)
df1['activation_month_cos'] = np.cos(2 * np.pi * df1['DCP_activation_date_month'] / 12)
df1['activation_day_sin'] = np.sin(2 * np.pi * df1['DCP_activation_date_day'] / 7)
df1['activation_day_cos'] = np.cos(2 * np.pi * df1['DCP_activation_date_day'] / 7)

df1['start_month_sin'] = np.sin(2 * np.pi * df1['DCP_start_date_month'] / 12)
df1['start_month_cos'] = np.cos(2 * np.pi * df1['DCP_start_date_month'] / 12)
df1['start_day_sin'] = np.sin(2 * np.pi * df1['DCP_start_date_day'] / 7)
df1['start_day_cos'] = np.cos(2 * np.pi * df1['DCP_start_date_day'] / 7)

df1['drop_month_sin'] = np.sin(2 * np.pi * df1['DCP_drop_date_month'] / 12)
df1['drop_month_cos'] = np.cos(2 * np.pi * df1['DCP_drop_date_month'] / 12)
df1['drop_day_sin'] = np.sin(2 * np.pi * df1['DCP_drop_date_day'] / 7)
df1['drop_day_cos'] = np.cos(2 * np.pi * df1['DCP_drop_date_day'] / 7)

df1.info()
```

```python
#drop raw dates
df1.drop(['DCP_activation_date','DCP_start_date','DCP_drop_date'],axis = 1,inplace=True)
```

```python
#dropping the rest
#drop all objects except drop status for random forest
#df1 = df1.select_dtypes(exclude=['object'])

df1 = df1.drop(columns=['postal_prefix_at_activation','city_at_activation','province_at_activation'])
```
